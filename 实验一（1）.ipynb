```python
import requests
from bs4 import BeautifulSoup
import re
import os
```

```python
#获取国家知识产权局统计年报首页内容
r = requests.get("http://www.sipo.gov.cn/tjxx/gjzscqjtjnb/index.htm")
print(r.status_code)
r.encoding = r.apparent_encoding
soup = BeautifulSoup(r.text,"lxml")
```

```python
#获取所有年份的首页ulr
urls = []
for url in soup.find_all('a',target = '_blank'):
    urls.append(url.get('href').replace(" ",""))
urls
```

```python
#获取各年专利统计年报的子页面中的url
sub_url_list=[]
for i in range(len(urls)):
    sub_r = requests.get(urls[i])
    sub_html = sub_r.text
    sub_soup = BeautifulSoup(sub_html,"lxml")
    sub_urls=[]
    for url in sub_soup.find_all('a',onmouseout=re.compile(r"m{2}_swapimgrestore()",re.I)):
        sub_url = os.path.dirname(urls[i]) + '/' +url.get('href')
        sub_urls.append(sub_url)
    sub_url_list.append(sub_urls)
sub_url_list
```

```python
#获取每个按钮对应的页面里的所有数据页面的url
a = dict()
for i in sub_url_list:
    _html=''
    for k in i:
        if 'a.html' not in k:
            continue
        else:
            _html = k
    if _html =='':
        continue
    r = requests.get(_html)
    a[i[0]]=[]
    r.encoding=r.apparent_encoding
    html = r.text
    soup = BeautifulSoup(html,"lxml")
    for url in soup.find_all("a",href = re.compile("/")):
        url = os.path.dirname(i[0]) + '/' +url.get('href')
        a[i[0]].append(url)
```

```python
alist = list(a.values())
alist
```

```python
import pandas as pd
#获取表格，存入excel
def run():
    for year in range(2018,1984,-1):
        n = 0
        for url in alist[2018-year]:        
            tb = pd.read_html(url)[3]  #经观察发现所需表格是网页中第4个表格，故为[3]
            tb.to_excel('{}年{}表.xls'.format(year,n), encoding='utf_8_sig', header=1, index=0)
            print('{}年{}表抓取完成'.format(year,n))
            n = n+1
        print('{}年数据全部抓取完成！'.format(year)) 
#这里爬取2018年的数据作为展示
year = 2018
n = 0
for url in alist[2018-year]:        
    tb = pd.read_html(url)[3]  #经观察发现所需表格是网页中第4个表格，故为[3]
    tb.to_excel('{}年{}表.xls'.format(year,n), encoding='utf_8_sig', header=1, index=0)
    print('{}年{}表抓取完成'.format(year,n))
    n = n+1
print('{}年数据全部抓取完成！'.format(year)) 
```
