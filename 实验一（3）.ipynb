# 三、从佰腾网完成各省不同类型专利数据的收集

```python
import requests
import json
import pandas as pd
```

```python
data=[]
provs = ['北京','天津','上海','重庆','河北','山西','辽宁','吉林','黑龙江','江苏','浙江','安徽','福建','江西','山东','河南','湖北','湖南','广东','海南','四川','贵州','云南','陕西','甘肃','青海','台湾','内蒙古','广西','西藏','宁夏','新疆','香港','澳门']
for prov in provs:    #按省份循环爬取
    for year in range(1985,2020):   #按年份循环爬取
        #进行request请求
        headers = {
            'origin': 'https://www.baiten.cn',
            'user-agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36',
        }
        params={
            'q': 'aa:({0}) AND (pd:({1}))'.format(prov,year),
        }
        url = 'https://www.baiten.cn/results/filter'
        res_patent = requests.post(url,headers=headers,params=params)
        #获取需要的数据
        try:
            json_patent = res_patent.json()
            dict_patent = json_patent['facetPivots']['country']['cn']
            dict_patent['年份']='{}'.format(year)
            dict_patent['省份']='{}'.format(prov)
            data.append(dict_patent)     
        except:
            continue
    print("{}数据爬取完成".format(prov))
print("数据全部爬取完成！！！")
data = pd.DataFrame(data).fillna(0)   #将Nan值转化为0
#修改列名，将浮点型数据转换为整数型,规定列的顺序
data.rename(columns={'cn_um':'实用新型专利', 'cn_in':'发明专利', 'cn_id':'外观专利', 'cn_gp':'发明授权专利', 'cn_in_gp':'发明公开专利'}, inplace = True)
data[['发明专利','发明公开专利','发明授权专利','外观专利','实用新型专利']] = data[['发明专利','发明公开专利','发明授权专利','外观专利','实用新型专利']].astype(int)
#存入excel
data.to_excel('Patent.xls',index = False)
```
